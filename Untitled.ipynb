```python
df11 = pd.read_csv('data/tn.movie_budgets.csv')
df11['domestic_gross'] = [x[1:].replace(',', '') for x in df11['domestic_gross']]
df11['domestic_gross'] = [int(x, base=10) for x in df11['domestic_gross']]
df11.sort_values(['domestic_gross'], ascending=False)
```

```python
merged_df = df5.merge(df6, left_on=['title_id'], right_on=['tconst'])
merged_df
```

```python
merged_df2 = merged_df.merge(df7, on=['tconst'])
merged_df2
```

```python
merged_df3 = merged_df2.merge(df8, on=['tconst'])
merged_df3
```

```python
merged_df4 = merged_df3.merge(df9, on=['tconst'])
merged_df4[merged_df4['region'] == 'US']
```

```python
title_group = merged_df4.groupby(['primary_title', 'genres']).mean()
title_group = title_group.sort_values(['numvotes'], ascending=False)
```

```python
title_group = title_group[title_group['numvotes'] >= 5000]
title_group = title_group[title_group['averagerating'] >= 7.5]
title_group = title_group.sort_values(['averagerating'], ascending=False)
```

```python
condensed_df = title_group.drop(['is_original_title', 'ordering_x', 'ordering_y'], 'columns')
```

```python
condensed_df
```

```python
condensed_df = condensed_df.merge(df1, left_on=['primary_title'], right_on=['title']).groupby(['title'])
```
