## What genres perform best at the box office?

```python
import pandas as pd
```

```python
# Make a DataFrame containing information about genres and gross sales.
title_basics = pd.read_csv('data/title.basics.csv')
gross = pd.read_csv('data/bom.movie_gross.csv')
genres_df = title_basics.merge(gross, left_on=['primary_title'], right_on=['title'])
```

```python
# Drop columns to make the DataFrame easier to look at.
genres_df = genres_df.drop(columns=['tconst', 'primary_title', 'original_title', 'runtime_minutes', 'start_year'])
```

```python
# Convert genres column to string so NaN values don't mess up the next step.
genres_df['genres'] = genres_df['genres'].astype(str)
```

```python
# New dataframe where we split the genres column and restack to get one genre per row.
df1 = pd.DataFrame({'Title': [x for x in genres_df['title']], 'Genre': [x for x in genres_df['genres']]})
new_df = pd.DataFrame(df1.Genre.str.split(',').tolist(), index=df1.Title).stack()
```

```python
# Setting index and columns of new dataframe to get it ready to merge.
new_df = new_df.reset_index([0, 'Title'])
new_df.columns = ['Title', 'Genre']
```

```python
# Drop duplicate rows.
new_df = new_df.drop_duplicates()
```

```python
# Merge new dataframe into original.
genres_df = genres_df.merge(new_df, left_on='title', right_on='Title')
```

```python
# Drop original multi-genre column since we don't need it anymore.
genres_df = genres_df.drop(columns=['genres'])
```

```python
genres_df.groupby(['year', 'Genre']).mean().sort_values(['year', 'domestic_gross'], ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plot showing yearly domestic gross by genre.
fig, ax1 = plt.subplots(figsize=(15,10))
sns.lineplot(x=genres_df['year'], y=genres_df['domestic_gross'], hue=genres_df['Genre'], ax=ax1, palette='Accent')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/rt.movie_info.tsv', sep='\t')
df['box_office'] = df['box_office'].astype(str)
df['box_office'] = [x.replace(',','') for x in df['box_office']]
# df['box_office'] = [x.replace('nan', '-9999') for x in df['box_office']]
df['box_office'] = df['box_office'].astype(float, errors='ignore')
df.sort_values(['box_office'], ascending=False)
```

```python
df.groupby(['director', 'id']).mean().sort_values(['box_office'], ascending=False)
```

```python
import numpy as np
(directors, counts) = np.unique(df['director'].astype(str), return_counts=True)
freq = np.asarray((directors, counts)).T
freq
```

```python
maybe = pd.DataFrame(freq)
maybe = maybe.rename(columns={0: 'director', 1: 'dir_count'})
```

```python
df = df.merge(maybe, on=['director'])
```

```python
df = df.drop(columns=['synopsis', 'currency', 'id'])
```

```python
df.groupby(['dir_count', 'director']).mean().sort_values(['dir_count', 'box_office'], ascending=False).head(50)
```
